# Grey Color Scale

```python
import pandas as pd

from lets_plot import *
LetsPlot.setup_html()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv')
```

```python
p = ggplot(df, aes(x='fl')) + geom_bar(aes(color='fl', fill='fl'), alpha=.5)
p1 = p + ggtitle("Default")
p2 = p + scale_color_grey(start=0, end=.7) + \
         scale_fill_grey(start=0, end=.7) + \
         ggtitle("With Grey Color Scale")

w, h = 400, 300
bunch = GGBunch()
bunch.add_plot(p1, 0, 0, w, h)
bunch.add_plot(p2, w, 0, w, h)
bunch
```
