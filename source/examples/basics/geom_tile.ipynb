# geom_tile()

```python
import numpy as np
from scipy.stats import multivariate_normal

from lets_plot import *
LetsPlot.setup_html()
```

```python
np.random.seed(42)
n = 25
x = np.linspace(-1, 1, n)
y = np.linspace(-1, 1, n)
X, Y = np.meshgrid(x, y)
mean = np.zeros(2)
cov = [[1, .5],
       [.5, 1]]
rv = multivariate_normal(mean, cov)
Z = rv.pdf(np.dstack((X, Y)))
data = {'x': X.flatten(), 'y': Y.flatten(), 'z': Z.flatten()}
```

```python
ggplot(data, aes('x', 'y')) + geom_tile(aes(fill='z'))
```
