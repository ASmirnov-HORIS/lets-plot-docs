# geom_linerange()

```python
import pandas as pd

from lets_plot import *
LetsPlot.setup_html()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv')
class_df = df.groupby('class').hwy.agg(['min', 'median', 'max']).reset_index()
```

```python
ggplot(class_df, aes(x='class')) + geom_linerange(aes(ymin='min', ymax='max'))
```
