# Manual Shape Scale

```python
import pandas as pd

from lets_plot import *
LetsPlot.setup_html()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv')
fl_df = df.groupby(['cty', 'hwy']).fl.agg(pd.Series.mode).to_frame('fl').reset_index()
```

```python
p = ggplot(fl_df, aes('cty', 'hwy')) + geom_point(aes(shape='fl'))
p1 = p + ggtitle("Default")
p2 = p + scale_shape_manual(values=[0, 12, 1, 10, 3, 13, 2, 4]) + \
         ggtitle("With Manual Shape Scale")

w, h = 400, 300
bunch = GGBunch()
bunch.add_plot(p1, 0, 0, w, h)
bunch.add_plot(p2, w, 0, w, h)
bunch
```
